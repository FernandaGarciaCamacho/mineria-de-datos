##### Mineria de datos



Actividad para extracción de datos, donde se busca encontrar patrónes de datos para a su vez optimizar actividades. Se utiliza para procesar y explorar grandes conjuntos de datos. Es el proceso de analizar grandes volúmenes de datos para descubrir patrones, anomalías y correlaciones ocultas, transformando así datos brutos en conocimiento valioso para la toma de decisiones estratégicas, la predicción de tendencias y la optimización de procesos en diversas industrias, utilizando técnicas de estadística, inteligencia artificial y aprendizaje automático.

##### Ciencia de Datos

La ciencia de datos es el estudio de datos con el fin de extraer información significativa. Es el estudio científico de los datos para obtener conocimientos. La Ciencia de Datos es un campo multidisciplinario que combina estadística, matemáticas, inteligencia artificial y computación para analizar grandes volúmenes de datos y extraer conocimiento significativo, patrones y tendencias, con el fin de apoyar la toma de decisiones, hacer predicciones y resolver problemas en diversas industrias.

Media (Mean)

```python
data = np.array([1, 2, 3, 4, 5])
media = np.mean(data)
print(media)
```

Promedio (Average) y Promedio Ponderado

```python
data = np.array([1, 2, 3, 4, 5])
weights = np.array([1, 2, 3, 4, 5])
promedio = np.average(data)
promedio_ponderado = np.average(data, weights=weights)
```

Mediana

```python
mediana = np.median(data)
```

Varianza y Desviación Estándar

```python
varianza = np.var(data)
desviacion = np.std(data)
```

Generación de Datos Aleatorios

```python
muestra = np.random.normal(0, 1, 1000)
print(np.mean(muestra), np.std(muestra))
```

Correlación
